```python
weather_data['Date'] = weather_data['Date'].apply(pd.to_datetime)
```

```python
weather_data['Year'] = weather_data['Date'].dt.year
weather_data['Month'] = weather_data['Date'].dt.month
weather_data['Day'] = weather_data['Date'].dt.day
```

```python
weather_data.drop('Date', axis=1, inplace=True)
```

```python
for col in num_cols:
    col_median = weather_data[col].median()
    weather_data[col].fillna(col_median, inplace=True)

weather_data[num_cols].describe()
```

```python
cat_cols.remove('Date')
weather_data[cat_cols].isna().sum()
```

```python
for col in cat_cols:
    weather_data[col].fillna(weather_data[col].mode()[0], inplace=True)

weather_data[cat_cols].describe()
```

```python
weather_data.shape
```

```python
cols_to_keep =[
    'Location',
    'Rainfall',
    'Evaporation',
    'Sunshine',
    'WindGustDir',
    'WindGustSpeed',
    'WindDir3pm',
    'AvgWindSpeed',
    'Humidity3pm',
    'Cloud3pm',
    'RainToday',
    'Pressure3pm',
    'TempDiff',
]
target_col = ['RainTomorrow']
```

```python
weather_data['TempDiff'] = weather_data['MaxTemp'] - weather_data['MinTemp']
weather_data['AvgWindSpeed'] = (weather_data['WindSpeed9am'] + weather_data['WindSpeed3pm']) / 2
```

### Preparing dataset for predictions :

```python
from sklearn.model_selection import train_test_split

X = weather_data[cols_to_keep]
y = weather_data[target_col]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
print(X_train.shape)
```

```python
num_cols = X.select_dtypes(np.number).columns.tolist()
cat_cols = X.select_dtypes('object').columns.tolist()
```

### Working with Pipelines :

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('std_scaler', StandardScaler()),
])

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', OneHotEncoder(), cat_cols),
])
```

```python
piped_X_train = full_pipeline.fit_transform(X_train)
piped_X_test = full_pipeline.fit_transform(X_test)
print(piped_X_train.shape)
print(piped_X_test.shape)
```

### Defining model and Testing :

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(solver='liblinear', random_state=42)
log_reg.fit(piped_X_train, y_train.values.ravel())
```

```python
from sklearn.metrics import accuracy_score

y_pred_train = log_reg.predict(piped_X_train)
print('Train:', accuracy_score(y_train, y_pred_train))

y_pred_test = log_reg.predict(piped_X_test)
print('Test:', accuracy_score(y_test, y_pred_test))
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score

print('Recall :', recall_score(y_test, y_pred_test, average="binary", pos_label="No"))
print('Precision:', precision_score(y_test, y_pred_test, average="binary", pos_label="No"))
print('F1_score:', f1_score(y_test, y_pred_test, average="binary", pos_label="No"))
```
